```python
df = pd.read_csv('Data-Collisions.csv')
df.head()
```

```python
df.columns
df_needed = df[['SEVERITYCODE','PERSONCOUNT','VEHCOUNT', 'JUNCTIONTYPE', 'WEATHER', 'ROADCOND', 'LIGHTCOND']]
df_needed.head()
```

```python
df_needed = df_needed.dropna()
df_needed
print(df_needed.JUNCTIONTYPE.unique())
print(df_needed.WEATHER.unique())
print(df_needed.ROADCOND.unique())
print(df_needed.LIGHTCOND.unique())
```

```python
df_needed['JUNCTIONTYPE'] = df_needed['JUNCTIONTYPE'].astype('category')
df_needed['JUNCTIONTYPE'] = df_needed['JUNCTIONTYPE'].cat.codes
df_needed
```

```python
df_needed['WEATHER'] = df_needed['WEATHER'].astype('category')
df_needed['WEATHER'] = df_needed['WEATHER'].cat.codes
df_needed['ROADCOND'] = df_needed['ROADCOND'].astype('category')
df_needed['ROADCOND'] = df_needed['ROADCOND'].cat.codes
df_needed['LIGHTCOND'] = df_needed['LIGHTCOND'].astype('category')
df_needed['LIGHTCOND'] = df_needed['LIGHTCOND'].cat.codes
df_needed
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import tree
y_data = df_needed['SEVERITYCODE'].values
x_data = df_needed.drop('SEVERITYCODE', axis=1).values
x_data = preprocessing.StandardScaler().fit(x_data).transform(x_data.astype(float))
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.20, random_state=1)
print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
from sklearn import metrics
clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth = 5)
clf.fit(x_train,y_train)
predTree = clf.predict(x_test)
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, predTree))
```

```python
import pydotplus
import collections
data_feature_names = ['PERSONCOUNT','VEHCOUNT', 'JUNCTIONTYPE', 'WEATHER', 'ROADCOND', 'LIGHTCOND']
dot_data = tree.export_graphviz(clf,
                                feature_names=data_feature_names,
                                out_file=None,
                                filled=True,
                                rounded=True)
graph = pydotplus.graph_from_dot_data(dot_data)

colors = ('turquoise', 'orange')
edges = collections.defaultdict(list)

for edge in graph.get_edge_list():
    edges[edge.get_source()].append(int(edge.get_destination()))

for edge in edges:
    edges[edge].sort()    
    for i in range(2):
        dest = graph.get_node(str(edges[edge][i]))[0]
        dest.set_fillcolor(colors[i])

graph.write_png('tree.png')
```
